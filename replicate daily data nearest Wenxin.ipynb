```python
import numpy as np
import pandas as pd
import csv
import math
```

```python
data1=pd.read_csv("UnderlyingOptionsEODCalcs_2018-02-21.csv")
```

```python
nearg=data1[data1["daytoexp"]==30.0]
```

```python
nextg=data1[data1["daytoexp"]==36.0]
```

```python
R=0.0139# nextg
T1 = (60*7.75+60*16+1440*29)/525600
T2 = (60*7.75+60*16+1440*35)/525600
# print(T1,T2)
```

```python
nearcall=nearg[nearg["option_type"]=="c"]
nearcall.head()
nearcall["ave"]=(nearcall["bid_eod"]+nearcall["ask_eod"])/2
nearcall.head()
```

```python
nearput=nearg[nearg["option_type"]=="p"]
nearput.head()
nearput["ave"]=(nearput["bid_eod"]+nearput["ask_eod"])/2
nearput.head()
```

```python
pave=nearput[["ave","strike"]]
```

```python
near=pd.merge(nearcall,pave,on="strike")
near.head()
```

```python
near["difference"]=near["ave_x"]-near["ave_y"]
near["abs_difference"]=abs(near["difference"])
```

```python
mindiff=min(near["abs_difference"])
strikeline=near.loc[near["abs_difference"]==mindiff]
strike1=strikeline["strike"]
absdiff1=strikeline['difference']
F1=strike1+math.exp(R*T1)*absdiff1
F1
#because risk free rate==0,so e**(r*t)==1
```

```python
nextcall=nextg[nextg["option_type"]=="c"]
nextcall.head()
nextcall["ave"]=(nextcall["bid_eod"]+nextcall["ask_eod"])/2
nextput=nextg[nextg["option_type"]=="p"]
nextput.head()
nextput["ave"]=(nextput["bid_eod"]+nextput["ask_eod"])/2
pave1=nextput[["ave","strike"]]
nex=pd.merge(nextcall,pave1,on="strike")
nex.head()
nex["difference"]=nex["ave_x"]-nex["ave_y"]
nex["abs_difference"]=abs(nex["difference"])
```

```python
mindif=min(nex["abs_difference"])
strikel=nex.loc[nex["abs_difference"]==mindif]
strike2=strikel["strike"]
absdiff2=strikel['difference']
F2=strike2+math.exp(R*T2)*absdiff2
F2
```

```python
nearput["strike-1"]=nearput["strike"].shift(1)
nearput["strike1"]=nearput["strike"].shift(-1)
nearput["delk"]=(nearput["strike1"]-nearput["strike-1"])/2
# nearput.iloc[0]["delk"]=(nearput.iloc[0]["strike1"]-nearput.iloc[0]["strike"])/2
nearput["part"]=nearput["delk"]/(nearput["strike"]**2)*nearput["ave"]
nearput

sum1=np.sum(nearput["part"])
sum1=sum1+(nearput.iloc[0]["strike1"]-nearput.iloc[0]["strike"])/2/(nearput.iloc[0]["strike"]**2)*nearput.iloc[0]["ave"]
sum1=sum1+(nearput.iloc[-1]["strike"]-nearput.iloc[-1]["strike-1"])/2/(nearput.iloc[-1]["strike"]**2)*nearput.iloc[-1]["ave"]
```
